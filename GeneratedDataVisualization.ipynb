```python
import pandas as pd
import numpy as np
import plotly.express as px

data_augm = pd.read_csv("GeneratedData.csv")
#A tf.Tensor represents a multidimensional array of elements.
data1 = np.asarray(data_augm)
data = np.zeros((159,2048,3))
for i in range(data.shape[0]):
  x= data1[i,:]
  for j in range(2048):
    str1 = x [j] 
    a = str1.split('(')
    b = a[1].split(',')
    m = np.array(np.matrix(b[0])).ravel()
    data[i,j,:] = m


categ = {}
categ[0] = "bathtub"
categ[1] = "bed"
categ[2] = "sofa"
categ[3] = "dresser"
categ[4] = "night_stand"
categ[5] = "table"
categ[6] ="chair"
categ[7] = "desk"
categ[8] = "monitor"
categ[9] = "toilet"


i=0
while(i < 160):
  points=data[i,:,:]
  fig = px.scatter_3d(points,points[:, 0], points[:, 1], points[:, 2])
  fig.update_traces(mode='markers',marker=dict(size=3,color='grey',opacity=1,line=dict(width=0.01,color='DarkSlateGrey')))
  print("================",i)
  fig.show()
  if i<16:
    fileno =categ[0]+"-"+str(i)
  elif i>=16 and i<32:
    fileno =categ[1]+"-"+str(i)
  elif i>=32 and i<48:
    fileno =categ[2]+"-"+str(i)
  elif i>=48 and i<64:
    fileno =categ[3]+"-"+str(i)
  elif i>=64 and i<80:
    fileno =categ[4]+"-"+str(i)
  elif i>=80 and i<96:
    fileno =categ[5]+"-"+str(i)
  elif i>=96 and i<112:
    fileno =categ[6]+"-"+str(i)
  elif i>=112 and i<128:
    fileno =categ[7]+"-"+str(i)
  elif i>=128 and i<144:
    fileno =categ[8]+"-"+str(i)
  else:
    fileno =categ[9]+"-"+str(i)
  fig.write_html("results/{0}.html".format(fileno))
  i=i+8
  
```
